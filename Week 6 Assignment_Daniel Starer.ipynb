```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
from scipy.stats import mannwhitneyu
```

```python
#Start of Chapter 11 Stories
#Story 1: Evaluating the Effectiveness of a Stress Reduction Program Using the Sign Test
```

In this story, a researcher wants to evaluate whether or not a stress reduction program affects participants' stress levels.  Since the the question regarding this data is whether or not the median change in stress levels are significantly different than zero, our alternative hypothesis is that the median difference between the two datasets is not zero.  The null hypothesis would be that the median difference between the two groups is zero.

```python
#Stress levels before and after participating in the program out of 100
before = np.array([85, 90, 78, 92, 88, 84, 91, 87, 89, 93, 85, 90, 78, 92, 88, 84, 91, 87, 89, 93])
after = np.array([80, 85, 75, 88, 83, 79, 85, 82, 84, 88, 80, 85, 75, 88, 83, 79, 85, 82, 84, 88])

#Performs the Sign Test using the Wilcoxon signed-rank test
statistic, p_value = wilcoxon (before, after)

#Prints the result
print(f"Test Statistic : {statistic}")
print(f"P-value: {p_value}")

#Interprets the results
if p_value < 0.05:
 display("We reject the null hypothesis: There is a significant difference in stress levels before and after completing the program.")
else:
 display("We fail to reject the null hypothesis: There is no significant difference in stress levels before and after completing the program.")
```

```python
#Story 2: Evaluating the Effectiveness of a Sleep Improvement Program Using the Sign Test
```

In this story, a researcher is evaluating whether or not sleep quality is improved after participants complete a sleep study program.  The data like the last story is on a scale of 0 to 100.  Data is not normally distributed, so again, we are testing to see whether or not the median change in sleep quality is significantly different than zero.  The alternative hypothesis would therefore be that the median difference between the two datasets is not zero and the null hypothesis would be that the median difference is zero.

```python
#Sleep quality before and after participating in the sleep study out of 100
before = np.array([60, 65, 55, 70, 68, 62, 66, 63, 59, 64, 60, 65, 55, 70, 68, 62, 66, 63, 59, 64])
after = np.array([72, 75, 65, 80, 78, 70, 76, 73, 69, 74, 72, 75, 65, 80, 78, 70, 76, 73, 69, 74])

#Performs the Sign Test using the Wilcoxon signed-rank test
statistic, p_value = wilcoxon (before, after)

#Prints the result
print(f"Test Statistic : {statistic}")
print(f"P-value: {p_value}")

#Interprets the results
if p_value < 0.05:
 display("We reject the null hypothesis: There is a significant difference in sleep quality before and after completing the program.")
else:
 display("We fail to reject the null hypothesis: There is no significant difference in sleep quality before and after completing the program.")
```

```python
#Story 3: Evaluating the Effectiveness of a Physical Fitness Program Using the Wilcoxon Rank-Sum Test
```

In this story, a researcher wants to evaluate whether participating in a physical fitness program significantly improves fitness levels compared to not participating.  Once again, the data is on a scale of 0 to 100 and data may not be distributed normally.  In this case, since a Wilcoxon Rank-Sum test is needed, we will determine if the data comes from the same distribution or not.  Our alternative hypothesis would be then that the data comes from two different distributions and our null hypothesis would be that it comes from the same distribution.

```python
#Fitness levels for two groups of participants out of 100
group1 = np.array([50, 55, 53, 60, 58, 50, 55, 53, 60, 58]) #Fitness levels with no fitness program
group2 = np.array([75, 80, 85, 78, 75, 80, 85, 78, 82 , 82]) #Fitness levels after participating in the fitness program

#Performs the Wilcoxon Rank-Sum test
statistic, p_value = mannwhitneyu (group1, group2)

#Prints the result
print(f"Test Statistic : {statistic}")
print(f"P-value: {p_value}")

#Interprets the results
if p_value < 0.05:
 display("We reject the null hypothesis: The distributions of the two groups are different.")
else:
 display("We fail to reject the null hypothesis: The distributions between the two groups are the same")
```

```python
#Start of Chapter 13 Stories
#Story 1: E-commerce Product Recommendation Using Bayes’ Rule
```

```python
#Defines the likelihood of each known value
P_Purchase = 0.2  #Prior probability
P_View_given_Purchase = 0.6  #Likelihood: P(View | Purchase)
P_View_given_Not_Purchase = 0.3  #Likelihood: P(View | ¬Purchase)
P_Not_Purchase = 1 - P_Purchase  #Complement of Purchase

#Calculates the total probability of viewing the product
P_View = (P_View_given_Purchase * P_Purchase) + (P_View_given_Not_Purchase * P_Not_Purchase)

#Calculates the posterior probability using Bayes' Rule
P_Purchase_given_View = (P_View_given_Purchase * P_Purchase) / P_View

#Plots the output
labels = ["Prior likelihood", "View and buying", "View and not buying", "Posterior likelihood"]
values = [P_Purchase, P_View_given_Purchase, P_View_given_Not_Purchase, P_Purchase_given_View]

plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['blue', 'green', 'red', 'purple'])
plt.ylim(0, 1)
plt.ylabel("Probability")
plt.title("Bayesian Inference for viewing a product and purchasing it")
plt.show()

#Prints the results
print(f"Prior Probability P(Purchase): {P_Purchase:.2f}")
print(f"Likelihood P(View | Purchase): {P_View_given_Purchase:.2f}")
print(f"Likelihood P(View | ¬Purchase): {P_View_given_Not_Purchase:.2f}")
print(f"Total Probability P(View): {P_View:.2f}")
print(f"Posterior Probability P(Purchase | View): {P_Purchase_given_View:.2f}")
```

After viewing the product, as we can see with this bar graph, the overall likelihood of buying the product increased on average.  The prior likelihood started out at 20% but jumped to about 36% for the posterior likelihood.  This would seem to indicate that viewing the product did have an impact on a potential customer's decision to purchase the product.  Indeed, that is reflected in 60% of users who had viewed the product and decided to purchase it.  However, it is also of note that "viewed but didn't buy" is close to the value of the posterior likelihood.  It would be worth investigating this to determine what about the product those users didn't like or didn't find valuable when viewing the product.

```python
#Story 2: Online Streaming Subscription Using Bayes’ Rule
```

```python
#Defines the likelihood of each known value
P_Subscribe = 0.15  #Prior probability
P_Watch_given_Subscribe = 0.5  #Likelihood: P(Watch | Subscribe)
P_Watch_given_Not_Subscribe = 0.25  #Likelihood: P(Watch | ¬Subscribe)
P_Not_Subscribe = 1 - P_Subscribe  #Complement of Subscribe

#Calculates the total probability of watching the movie
P_Watch = (P_Watch_given_Subscribe * P_Subscribe) + (P_Watch_given_Not_Subscribe * P_Not_Subscribe)

#Calculates the posterior probability using Bayes' Rule
P_Subscribe_given_Watch = (P_Watch_given_Subscribe * P_Subscribe) / P_Watch

#Plots the output
labels = ["Prior likelihood", "Subscribed and watched", "Watched but did not subscribe", "Posterior likelihood"]
values = [P_Subscribe, P_Watch_given_Subscribe, P_Watch_given_Not_Subscribe, P_Subscribe_given_Watch]

plt.figure(figsize=(7, 5))
plt.bar(labels, values, color=['blue', 'green', 'red', 'purple'])
plt.ylim(0, 1)
plt.ylabel("Probability")
plt.title("Bayesian Inference for watching a movie and subscribing")

#Rotates labels for better readability
plt.xticks(rotation=30, ha="right")
plt.show()

#Prints the results
print(f"Prior P(Subscribe): {P_Subscribe:.3f}")
print(f"Likelihood P(Watch | Subscribe): {P_Watch_given_Subscribe:.3f}")
print(f"Likelihood P(Watch | No Subscribe): {P_Watch_given_Not_Subscribe:.3f}")
print(f"Total P(Watch): {P_Watch:.3f}")
print(f"Posterior P(Subscribe | Watch): {P_Subscribe_given_Watch:.3f}")
```

In this data, we can see the prior likelihood of watching the show before subscribing is rather low, and although the likelihood did increase in the posterior likelihood, the posterior value (at aroudn 26%) is very close to the value for "watched but did not subscribe" at 25%.  This could indicate that users don't feel the need to subscribe to the premium plan to watch the movie or have watched it already from another source.
